## Connect to STAC Catalog

```python

# Install dependencies (uncomment if not installed)
# !pip install pystac-client

import pystac_client

# Open the STAC catalog
url = "https://catalog.dive.edito.eu"
client = pystac_client.Client.open(url)
print("STAC Client:", client)
```

## List all collections

```python
collections = list(client.get_collections())
print(f"Found {len(collections)} collections")
```

## Filter collections by variable name

```python

variable = "human_activities"

matching_collections = [c for c in collections if variable in c.id]

print(f"Collections containing '{variable}':")
for c in matching_collections:
    print("•", c.id)
```

## Explore items in matching collections

```python
products = []

for collection in matching_collections:
    print(f"\n=== Collection: {collection.id} ===")
    for i, item in enumerate(collection.get_items()):
        print(f"Item {i}: {item.id}")
        print("  Datetime:", item.datetime)
        print("  Product Identifier:", item.properties.get("productIdentifier", "N/A"))
        print("  Available assets:", list(item.assets.keys()))
        for asset_key, asset in item.assets.items():
            print(f"    - {asset_key}: {asset.href[:100]}...")
        products.append(item.id)

print(f"\nTotal number of products found: {len(products)}")
```
